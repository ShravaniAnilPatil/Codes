```
import pandas as pd

# Assuming df is your original DataFrame
df_encoded = df.copy()  # Create a copy to keep the original data intact

# Apply One-Hot Encoding to the 'Genres' column
df_encoded = df_encoded.join(df_encoded['Genres'].str.get_dummies(sep=', '))

# Drop the original 'Genres' column as it's now encoded
df_encoded.drop(columns=['Genres'], inplace=True)

# Save the new dataset to a CSV file (optional)
df_encoded.to_csv('movies_encoded.csv', index=False)

# Display the first few rows to verify
print(df_encoded.head())
```

```
df.drop(columns=['Production_Countries'], inplace=True)
```

```
import numpy as np

# Calculate Q1 (25th percentile) and Q3 (75th percentile)
Q1 = df_encoded['Year'].quantile(0.25)
Q3 = df_encoded['Year'].quantile(0.75)

# Compute IQR
IQR = Q3 - Q1

# Define lower and upper bounds for outliers
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Find outliers
outliers = df_encoded[(df_encoded['Year'] < lower_bound) | (df_encoded['Year'] > upper_bound)]

print("Outliers in Release Year:")
print(outliers[['Year']])
```
