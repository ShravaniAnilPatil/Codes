```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df=pd.read_csv('movies_encoded.csv')
df.head()
```

```python
 # Printing first 15 values from dataset
print(df.head(15))
```

```python
 # Description of data
 print(df.describe())
```

Computing correlation matrix

```python
# Select numerical columns for correlation analysis
numeric_columns = ['$Worldwide', '$Domestic', '$Foreign', 'Domestic %', 'Foreign %', 'Year', 'Rating', 'Vote_Count']

# Compute correlation matrix
correlation_matrix = df[numeric_columns].corr()

# Print correlation matrix
print(correlation_matrix)
```

Calculating mean , median , mode

```python
# Calculate mean for key numerical attributes
mean_values = df[['$Worldwide', 'Rating', 'Vote_Count']].mean()
print(mean_values)
```

```python
# Calculate median for key numerical attributes
median_values = df[['$Worldwide', 'Rating', 'Vote_Count']].median()
print(median_values)
```

```python
# Calculate mode for key numerical attributes
mode_values = df[['Rating', 'Vote_Count', 'Language_Encoded']].mode()
print(mode_values)
```

Range

```python
# Compute Range for $Worldwide revenue and Rating
revenue_range = df['$Worldwide'].max() - df['$Worldwide'].min()
rating_range = df['Rating'].max() - df['Rating'].min()

print("Range of Worldwide Revenue:", revenue_range)
print("Range of Ratings:", rating_range)
```

IQR

```python
import numpy as np

# Compute Q1, Q3, and IQR
Q1_revenue, Q3_revenue = np.nanpercentile(df['$Worldwide'], [25, 75])
IQR_revenue = Q3_revenue - Q1_revenue

Q1_rating, Q3_rating = np.nanpercentile(df['Rating'], [25, 75])
IQR_rating = Q3_rating - Q1_rating

print("IQR of Worldwide Revenue:", IQR_revenue)
print("IQR of Ratings:", IQR_rating)
```

```python
# Compute Coefficient of Variation
# Compute Standard Deviation
revenue_std = df['$Worldwide'].std()
rating_std = df['Rating'].std()

print("Standard Deviation of Worldwide Revenue:", revenue_std)
print("Standard Deviation of Ratings:", rating_std)

cv_revenue = (revenue_std / df['$Worldwide'].mean()) * 100
cv_rating = (rating_std / df['Rating'].mean()) * 100

print("Coefficient of Variation for Worldwide Revenue:", cv_revenue, "%")
print("Coefficient of Variation for Ratings:", cv_rating, "%")
```

Calculating Standard deviation and variance for worldwide revenue

```python
# Standard Deviation
print("Standard Deviation of Worldwide Revenue:", df['$Worldwide'].std())

# Variance
print("Variance of Worldwide Revenue:", df['$Worldwide'].var())
```

Performing Chi Square Test for genre distrubution

```python
import scipy.stats

# Create a frequency table for genres
genre_freq = df.iloc[:, -10:].sum()  # Assuming last 10 columns are genre indicators (1 for presence, 0 for absence)

# Convert to DataFrame
genre_freq_df = pd.DataFrame(genre_freq, columns=['Count']).reset_index()
genre_freq_df.rename(columns={'index': 'Genre'}, inplace=True)

# Perform Chi-Square test
chi, p, df3, expected = scipy.stats.chi2_contingency(genre_freq_df['Count'].values.reshape(1, -1))
print("Chi-Square Value:", chi)
print("P-Value:", p)
```

Genre frequency table

Summary statisrics of Ratings

```python
import numpy as np

Q1, median, Q3 = np.nanpercentile(df['Rating'], [25, 50, 75])
min_rating, max_rating = df['Rating'].min(), df['Rating'].max()

print("Q1:", Q1, "Median:", median, "Q3:", Q3)
print("Minimum Rating:", min_rating, "Maximum Rating:", max_rating)
```

Visualizing the Correlation Using a Heatmap

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Set figure size
plt.figure(figsize=(10,6))

# Create heatmap
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")

# Title and show
plt.title("Correlation Heatmap of Movies Dataset")
plt.show()
```

Histogram â€“ Distribution of Movie Ratings

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set figure size
plt.figure(figsize=(10,6))

# Plot histogram
sns.histplot(df['Rating'], bins=20, kde=True, color='blue')

# Titles and labels
plt.title("Distribution of Movie Ratings", fontsize=14)
plt.xlabel("Movie Rating", fontsize=12)
plt.ylabel("Count of Movies", fontsize=12)

# Show plot
plt.show()
```

```python
Box Plot â€“ Domestic vs. Foreign Revenue Comparison
```

```python
plt.figure(figsize=(8,6))

# Plot boxplots
sns.boxplot(data=df[['$Domestic', '$Foreign']], palette="Set2")

# Titles and labels
plt.title("Domestic vs Foreign Revenue Distribution", fontsize=14)
plt.ylabel("Revenue (in billions)", fontsize=12)

# Show plot
plt.show()
```

Scatter Plot â€“ Relationship Between Vote Count & Revenue

```python
plt.figure(figsize=(10,6))

# Plot scatter plot
sns.scatterplot(x=df['Vote_Count'], y=df['$Worldwide'], alpha=0.6, color='red')

# Titles and labels
plt.title("Vote Count vs. Worldwide Revenue", fontsize=14)
plt.xlabel("Vote Count", fontsize=12)
plt.ylabel("Worldwide Revenue (in billions)", fontsize=12)

# Show plot
plt.show()
```

Bar Chart â€“ Most Popular Movie Genres

```python
# Count genres
genre_columns = ['Action', 'Comedy', 'Drama', 'Horror', 'Romance', 'Thriller', 'Science Fiction', 'War', 'Western']
genre_counts = df[genre_columns].sum().sort_values(ascending=False)

# Plot bar chart
plt.figure(figsize=(10,6))
sns.barplot(x=genre_counts.index, y=genre_counts.values, palette="coolwarm")

# Titles and labels
plt.title("Most Popular Movie Genres", fontsize=14)
plt.xlabel("Genre", fontsize=12)
plt.ylabel("Number of Movies", fontsize=12)
plt.xticks(rotation=45)

# Show plot
plt.show()
```
