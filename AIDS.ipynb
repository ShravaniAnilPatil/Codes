```python
import numpy as np
import pandas as pd
import scipy.stats as stats

train_data_numeric=train_data.select_dtypes(include=[np.number])
test_data_numeric=test_data.select_dtypes(include=[np.number])
mean_train=train_data_numeric.mean()
mean_test=test_data_numeric.mean()
std_train=train_data_numeric.std()
std_test=test_data_numeric.std()

n_train=train_data_numeric.shape[0]
n_test=test_data_numeric.shape[0]

sen_train=std_train/np.sqrt(n_train)
sen_test=std_test/np.sqrt(n_test)

z_statistic,p_value=stats.ttest_ind_from_stats(
    mean_train,sen_train,n_train,
    mean_test,sen_test,n_test,
    equal_var=False
)

if np.any(p_value<0.05):
  print("significant",p_value)
else:
  print("not")
```
