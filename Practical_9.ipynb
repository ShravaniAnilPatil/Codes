```python
cols =['Name', 'Ticket', 'Cabin']
df=df.drop(cols, axis=1)
print(df)
```

```python
dummies= []
cols = ['Pclass', 'Sex', 'Embarked']
for col in cols:
  dummies.append(pd.get_dummies(df[col]))

heart_dummies = pd.concat(dummies, axis=1)
df=pd.concat((df,heart_dummies), axis=1)
df.info()
```

```python
numeric_cols = df.select_dtypes(include='number').columns
df[numeric_cols] = df[numeric_cols].fillna(df[numeric_cols].mean())
print(df.shape)
print(df.isna().sum())
```

```python
df.drop(columns = ['Embarked'], inplace=True)
print(df.shape)
print(df.isna().sum())

```

```python
numeric_cols = df.select_dtypes(include='number').columns
Q1 = df[numeric_cols].quantile(0.25)
Q3 = df[numeric_cols].quantile(0.75)
IQR = Q3 - Q1
print(IQR)
```

```python
IQR["Fare"], Q1["Fare"] , Q3["Fare"]
```

```python
cut_off= IQR["Fare"] *1.5
lower, upper = Q1["Fare"] - cut_off, Q3["Fare"] + cut_off
print(df)
```

```python
df_x= df[["Fare", "Age"]]
df[["Fare", "Age"]]= (df_x-df_x.mean())/df_x.std()
df
```

```python
lower = Q1['Fare'] - 1.5 * IQR['Fare']
upper = Q3['Fare'] + 1.5 * IQR['Fare']

outliers = df[(df['Fare'] < lower) | (df['Fare'] > upper)]
outliers.head(10)
```

```python
outliers.shape
```

```python
df= pd.DataFrame([v for k, v in df.iterrows() if v["Fare"] >= lower or v["Fare"] <= upper])
```
